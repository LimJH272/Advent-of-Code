## Puzzle 1

```python
total1 = 0

nrows, ncols = grid.shape

for r in range(nrows):
    for c in range(ncols-3):
        curr = ''.join(grid[r, c:c+4])
        if curr in ('XMAS', 'SAMX'):
            total1 += 1

for r in range(nrows-3):
    for c in range(ncols):
        curr = ''.join(grid[r:r+4, c])
        if curr in ('XMAS', 'SAMX'):
            total1 += 1

for r in range(nrows-3):
    for c in range(ncols-3):
        row_indices = [r+i for i in range(4)]
        col_indices = [c+i for i in range(4)]
        curr = ''.join(grid[row_indices, col_indices])
        if curr in ('XMAS', 'SAMX'):
            total1 += 1

for r in range(3, nrows):
    for c in range(ncols-3):
        row_indices = [r-i for i in range(4)]
        col_indices = [c+i for i in range(4)]
        curr = ''.join(grid[row_indices, col_indices])
        if curr in ('XMAS', 'SAMX'):
            total1 += 1

total1
```
